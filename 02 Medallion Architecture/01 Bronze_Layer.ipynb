Bronze Layer :
Ingesting data from the 5 csv files that contains the raw files of the busniess.

01 Creating the bronze database 

```python
%sql
create database if not exists cash_flow_project.cash_flow_bronze
```

02 ingesting the files one by one and save it to a table

```python
file_path = "/Volumes/cash_flow_project/cash_flow_bronze/raw_data/checking_account_main.csv"
df = (spark.read
      .format("csv")
      .option("header", "true")
      .option("inferSchema", "true")
      .load(file_path))
(df.write
 .format("delta")
 .mode("overwrite")
 .saveAsTable("cash_flow_project.cash_flow_bronze.checking_accounts_main"))
```
